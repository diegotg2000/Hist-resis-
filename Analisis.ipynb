```python
N=500
L=0.072
A=0.0016

plt.figure(figsize=(9,6))

I=np.loadtxt('nucleoOx1.txt', usecols=0)
Phi=np.loadtxt('nucleoOx1.txt', usecols=1)

I=I-np.mean(I)
Phi=Phi-np.mean(Phi)

#H=NI/L Phi=NAB --> B=Phi/(NA)

H=N*I/L
B=Phi/(N*A)


plt.plot(B,H)
plt.xlabel('H (A/m)', fontsize=15)
plt.ylabel('B (T)', fontsize=15)
```

```python
N=500
L=0.072
A=0.0016

plt.figure(figsize=(9,6))

NOMBRES = np.array(['nucleoOx1.txt', 'nucleoOx2.txt', 'nucleoOx3.txt', 'nucleoOx4.txt'])

for nombre in NOMBRES:
    i=np.where(NOMBRES==nombre)[0][0]
    
    I=np.loadtxt(nombre, usecols=0)
    Phi=np.loadtxt(nombre, usecols=1)

    I=I-np.mean(I)
    Phi=Phi-np.mean(Phi)

    #H=NI/L Phi=NAB --> B=Phi/(NA)

    H=N*I/L
    B=Phi/(N*A)

    plt.subplot(2,2,i+1)
    plt.plot(B,H, label=nombre)
    plt.xlabel('H (A/m)', fontsize=15)
    plt.ylabel('B (T)', fontsize=15)
    plt.legend()
```
