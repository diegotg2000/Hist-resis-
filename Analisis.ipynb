```python
N=500
L=0.072
A=0.0016

plt.figure(figsize=(9,6))

NOMBRES = np.array(['nucleoOx1.txt', 'nucleoOx4.txt', 'nucleoOx2.txt', 'nucleoOx3.txt'])
DATOS = np.array(['f=0.1Hz, V=1V','f=0.2, V=1V','f=0.2Hz, V=2V','f=0.3Hz, V=2V'])
for nombre in NOMBRES:
    i=np.where(NOMBRES==nombre)[0][0]
    
    I=np.loadtxt(nombre, usecols=0)
    Phi=np.loadtxt(nombre, usecols=1)

    I=I-np.mean(I)
    Phi=Phi-np.mean(Phi)

    #H=NI/L Phi=NAB --> B=Phi/(NA)

    H=N*I/L
    B=Phi/(N*A)

    plt.subplot(2,2,i+1)
    plt.plot(H,B, label=DATOS[i], c='black')
    if i>1:
        plt.xlabel('H (A/m)', fontsize=15)
    if i%2 ==0:
        plt.ylabel('B (T)', fontsize=15)
    plt.axis([-4550,4550,-1,1])
    plt.legend()
```
